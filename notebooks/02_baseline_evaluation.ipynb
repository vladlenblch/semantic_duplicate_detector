```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT))
```

```python
import pandas as pd

from src.metrics import Metrics
from src.models.TFIDFBaseline import TFIDFBaseline
```

```python
train_df = pd.read_csv("../data/processed/train.csv")

train_df.head()
```

```python
val_df = pd.read_csv("../data/processed/val.csv")

val_df.head()
```

```python
model_on_logreg = TFIDFBaseline()
model_on_logreg.fit(train_df, mode="logreg")

y_pred_on_logreg = model_on_logreg.predict(val_df)
y_proba_on_logreg = model_on_logreg.predict_proba(val_df)
y_true = val_df["is_duplicate"]

logreg_report = Metrics.report(
    y_true=y_true,
    y_pred=y_pred_on_logreg,
    y_score=y_proba_on_logreg,
    k=0.1,
)

print(logreg_report)
```

```python
model_on_cosine = TFIDFBaseline()
model_on_cosine.fit(train_df, mode="cosine")

y_pred_on_cosine = model_on_cosine.predict(val_df)
y_proba_on_cosine = model_on_cosine.predict_proba(val_df)
y_true = val_df["is_duplicate"]

cosine_report = Metrics.report(
    y_true=y_true,
    y_pred=y_pred_on_cosine,
    y_score=y_proba_on_cosine,
    k=0.1,
)

print(cosine_report)
```
